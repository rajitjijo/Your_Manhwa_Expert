```python
website = "https://tbate.fandom.com/wiki/Characters"
```

```python
url = website
response = requests.get(url=website, headers=headers)

body = response.content

soup = BeautifulSoup(body, 'html.parser')

title = soup.title.string if soup.title else "Title Not Found"

if soup.body:

    for irrelevant in soup.body(["scrit", "stle", "img", "input", "nav", "figure"]):

        irrelevant.decompose()

    text = soup.body.get_text(separator="\n", strip=True)

else:

    text = ""

links = [link.get("href") for link in soup.find_all("a")]
links = [link for link in links if link]
```

```python
for tag in soup.find('div', class_='mw-parser-output').find_all(['h2', 'h3', 'ul']):
    print(tag.text)
```
